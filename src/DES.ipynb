**Some function**

```python
# Phép hoán vị bit
def permute(k, arr, n):
    permutation = ""
    for i in range(0, n):
        permutation = permutation + k[arr[i] - 1]
    return permutation

# Phép dịch trái nth bước
def shift_left(k, nth_shifts):
    s = ""
    for i in range(nth_shifts):
        for j in range(1, len(k)):
            s = s + k[j]
        s = s + k[0]
        k = s
        s = ""
    return k

# Phép XOR giữa hai chuỗi bit a và b
def xor(a, b):
    ans = ""
    for i in range(len(a)):
        if a[i] == b[i]:
            ans = ans + "0"
        else:
            ans = ans + "1"
    return ans

# Chuyển hex sang bin
def hex2bin(hexStr):
    # Up thành chữ viết hoa
    hexStr = hexStr.upper()
    # mapping
    mp = {'0': "0000", '1': "0001", '2': "0010", '3': "0011", 
          '4': "0100", '5': "0101", '6': "0110", '7': "0111",
          '8': "1000", '9': "1001", 'A': "1010", 'B': "1011",
          'C': "1100", 'D': "1101", 'E': "1110", 'F': "1111"}
    binStr = ""
    for i in range(len(hexStr)):
        binStr = binStr + mp[hexStr[i]]
    return binStr

# Chuyển bin sang hex
def bin2hex(binStr):
    mp = {"0000": '0', "0001": '1', "0010": '2', "0011": '3',
          "0100": '4', "0101": '5', "0110": '6', "0111": '7',
          "1000": '8', "1001": '9', "1010": 'A', "1011": 'B',
          "1100": 'C', "1101": 'D', "1110": 'E', "1111": 'F'}
    hexStr = ""
    # lấy lần lượt 4 bit/lần
    for i in range(0, len(binStr), 4):
        ch = ""
        ch = ch + binStr[i]
        ch = ch + binStr[i + 1]
        ch = ch + binStr[i + 2]
        ch = ch + binStr[i + 3]
        hexStr = hexStr + mp[ch]
    return hexStr

# Chuyển bin sang decimal
def bin2dec(binary):
    binary1 = binary
    decimal, i, n = 0, 0, 0
    while(binary != 0):
        dec = binary % 10
        decimal = decimal + dec * pow(2, i)
        binary = binary//10
        i += 1
    return decimal

# Chuyển decimal sang bin
def dec2bin(num):
    res = bin(num).replace("0b", "")
    if(len(res) % 4 != 0):
        div = len(res) / 4
        div = int(div)
        counter = (4 * (div + 1)) - len(res)
        for i in range(0, counter):
            res = '0' + res
    return res
```

**Hàm mã hóa**

```python
def keyGeneration(key):
    # Đưa hex về bin
    key = hex2bin(key)
    
    # Nếu độ dài là 64 thì đưa về 56 bằng pc1
    if len(key) == 64:
        key = permute(key, pc1, 56)
    
    # Chia thành C0 và D0
    left = key[0:28]
    right = key[28:56]
    
    subkeyList = [] # list các subkey dạng bin
    
    # lần lượt sinh 16 khóa con
    for i in range(0, 16):
        # Dịch bit sang trái n bước dựa vào shift table
        left = shift_left(left, shift_table[i])
        right = shift_left(right, shift_table[i])

        # Gộp hai phần lại
        combine_str = left + right

        # Đưa 56 bits về 48 bits bằng pc2
        subkey = permute(combine_str, pc2, 48)

        subkeyList.append(subkey)
    
    return subkeyList

def Feistel(subkey, right):
    # Expand right từ 32 bits sang 48 bits bằng expand box
    right_expanded = permute(right, expand_box, 48)
    
    # XOR subkey với right_expanded
    xor_x = xor(right_expanded, subkey)
    
    # Đưa vào 8 Sbox
    sbox_str = ""
    for j in range(0, 8):
        # hai bit đầu xác định row
        row = bin2dec(int(xor_x[j*6] + xor_x[j*6 + 5]))
        # bốn bit giữa xác định col
        col = bin2dec(int(xor_x[j * 6 + 1] + xor_x[j*6 + 2] + xor_x[j*6 + 3] + xor_x[j*6 + 4]))
        # giá trị ô sbox tương ứng
        val = sbox[j][row][col]
        # chuyển sang bin và nối vào kết quả
        sbox_str = sbox_str + dec2bin(val)
    
    # Đưa qua phép hoán vị Pbox (32 bits to 32 bits)
    sbox_str = permute(sbox_str, pbox, 32)
    
    return sbox_str
        
def encrypt(plainText, subkeyList):
    # Đưa hex về bin
    pt = hex2bin(plainText)
 
    # Phép hoán vị khởi đầu IP
    pt = permute(pt, initial_perm, 64)
    print("After initial permutation", bin2hex(pt))
 
    # Split thành L0 và R0
    left = pt[0:32]
    right = pt[32:64]
    
    # Đưa vòa 16 vòng lặp
    for i in range(0, 16):
        sbox_str = Feistel(subkeyList[i], right)
 
        # XOR left với sbox_str
        result = xor(left, sbox_str)
        left = result
 
        # Swap tại vòng lặp cuối
        if(i != 15):
            left, right = right, left
            
        # In thông tin từng loop
        print("Round ", i + 1, " ", bin2hex(left), " ", bin2hex(right), " ", bin2hex(subkeyList[i]))
 
    # Gộp L16 và R16
    combine = left + right
 
    # Phép hoán vị kết thúc FP
    cipher_text = permute(combine, final_perm, 64)
    return cipher_text
```
